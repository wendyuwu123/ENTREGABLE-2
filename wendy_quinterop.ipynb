```python
import numpy as np
import matplotlib.pyplot as plt

def mandelbrot(c, max_iter):
    z = 0
    for n in range(max_iter):
        if abs(z) > 2:
            return n
        z = z*z + c
    return max_iter



pixels = 800
max_iter = 100   # Aumenta detalle

center_x = -0.75   # Punto central real
center_y = 0.0     # Punto central imaginario
zoom = 1           # 1 = vista normal, >1 = más zoom


x_width = 3 / zoom
y_height = 3 / zoom

x = np.linspace(center_x - x_width/2, center_x + x_width/2, pixels)
y = np.linspace(center_y - y_height/2, center_y + y_height/2, pixels)

xx, yy = np.meshgrid(x, y)
c = xx + yy*1j


mandelbrot_set = np.array(
    [mandelbrot(ci, max_iter) for ci in c.ravel()]
).reshape(c.shape)



plt.figure(figsize=(8,8))
plt.imshow(mandelbrot_set,
           extent=[x.min(), x.max(), y.min(), y.max()])
plt.colorbar()
plt.title(f"Mandelbrot Set (zoom={zoom}, max_iter={max_iter})")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

def julia(z, c, max_iter):
    for n in range(max_iter):
        if abs(z) > 2:
            return n
        z = z*z + c
    return max_iter


# ==========================
# PARÁMETROS
# ==========================

pixels = 800
max_iter = 100
constante = -0.4 + 0.6j   # Puedes cambiarla

# Rango del plano complejo
x = np.linspace(-1.5, 1.5, pixels)
y = np.linspace(-1.5, 1.5, pixels)

xx, yy = np.meshgrid(x, y)
z = xx + yy*1j

# ==========================
# CÁLCULO
# ==========================

julia_set = np.array(
    [julia(zi, constante, max_iter) for zi in z.ravel()]
).reshape(z.shape)

# ==========================
# GRÁFICA
# ==========================

plt.figure(figsize=(8,8))
plt.imshow(julia_set,
           extent=[-1.5, 1.5, -1.5, 1.5])
plt.colorbar()
plt.title(f"Julia Set (c = {constante})")
plt.show()
```

```python
import cmath
import math
import matplotlib.pyplot as plt

def operaciones_basicas(z1, z2):
    suma = z1 + z2
    multiplicacion = z1 * z2
    modulo1 = abs(z1)
    modulo2 = abs(z2)
    conjugado1 = z1.conjugate()
    conjugado2 = z2.conjugate()

    print("Suma:", suma)
    print("Multiplicación:", multiplicacion)
    print("Módulo z1:", modulo1)
    print("Módulo z2:", modulo2)
    print("Conjugado z1:", conjugado1)
    print("Conjugado z2:", conjugado2)

    return suma  # devolvemos algo para poder graficarlo


def graficar_operacion(z1, z2, resultado, operacion):

    plt.figure(figsize=(10, 8))


    max_val = max(abs(z1.real), abs(z1.imag), abs(z2.real), abs(z2.imag),
                  abs(resultado.real), abs(resultado.imag)) + 1

    # Graficar vectores
    # Vector z1 (azul)
    plt.arrow(0, 0, z1.real, z1.imag, head_width=0.1*max_val/5,
              head_length=0.1*max_val/5, fc='blue', ec='blue', alpha=0.7,
              label=f'z₁ = {z1}')

    # Vector z2 (verde)
    plt.arrow(0, 0, z2.real, z2.imag, head_width=0.1*max_val/5,
              head_length=0.1*max_val/5, fc='green', ec='green', alpha=0.7,
              label=f'z₂ = {z2}')

    # Vector resultado (rojo)
    plt.arrow(0, 0, resultado.real, resultado.imag, head_width=0.1*max_val/5,
              head_length=0.1*max_val/5, fc='red', ec='red', alpha=0.9,
              linewidth=2, label=f'Resultado = {resultado}')

    # Configuración del gráfico
    plt.xlabel("Parte Real", fontsize=12)
    plt.ylabel("Parte Imaginaria", fontsize=12)
    plt.axhline(0, color='black', linewidth=1)
    plt.axvline(0, color='black', linewidth=1)
    plt.grid(True, alpha=0.3, linestyle='--')

    # Ajustar límites de los ejes
    plt.xlim(-max_val, max_val)
    plt.ylim(-max_val, max_val)

    plt.title(f"Operación: {operacion}", fontsize=14, fontweight='bold')
    plt.legend(loc='upper right')
    plt.gca().set_aspect('equal')  # Mantener proporción 1:1
    plt.tight_layout()
    plt.show()


def main():

    while True:

        x1 = float(input("Ingrese la parte real del primer número: "))
        y1 = float(input("Ingrese la parte imaginaria del primer número: "))

        x2 = float(input("Ingrese la parte real del segundo número: "))
        y2 = float(input("Ingrese la parte imaginaria del segundo número: "))


        z1 = complex(x1, y1)
        z2 = complex(x2, y2)

        while True:
            print("1. Suma ")
            print("2. Multiplicación ")
            print("3. Convertir z₁ a forma polar")
            print("4. Convertir z₂ a forma polar")
            print("5. Mostrar todas las operaciones básicas")
            print("6. Salir")

            opcion = int(input("Escoja una opción (1-7): "))
            if opcion == 1:
                resultado = z1 + z2
                print(f"\nResultado de la suma: {z1} + {z2} = {resultado}")
                graficar_operacion(z1, z2, resultado, "SUMA")

            elif opcion == 2:
                resultado = z1 * z2
                print(f"\nResultado de la multiplicación: {z1} × {z2} = {resultado}")
                graficar_operacion(z1, z2, resultado, "MULTIPLICACIÓN")

            elif opcion == 3:
                r, theta = cmath.polar(z1)
                print(f"\nForma polar de z₁ = {z1}:")
                print(f"  Magnitud (r): {r:.4f}")
                print(f"  Ángulo (radianes): {theta:.4f}")
                print(f"  Ángulo (grados): {math.degrees(theta):.2f}°")

            elif opcion == 4:
                r, theta = cmath.polar(z2)
                print(f"\nForma polar de z₂ = {z2}:")
                print(f"  Magnitud (r): {r:.4f}")
                print(f"  Ángulo (radianes): {theta:.4f}")
                print(f"  Ángulo (grados): {math.degrees(theta):.2f}°")

            elif opcion == 5:
                print("\n=== OPERACIONES BÁSICAS ===")
                suma = z1 + z2
                resta = z1 - z2
                multiplicacion = z1 * z2
                division = z1 / z2

                print(f"Suma: {z1} + {z2} = {suma}")
                print(f"Resta: {z1} - {z2} = {resta}")
                print(f"Multiplicación: {z1} × {z2} = {multiplicacion}")
                print(f"División: {z1} ÷ {z2} = {division}")
                print(f"Módulo de z₁: |{z1}| = {abs(z1):.4f}")
                print(f"Módulo de z₂: |{z2}| = {abs(z2):.4f}")
                print(f"Conjugado de z₁: {z1.conjugate()}")
                print(f"Conjugado de z₂: {z2.conjugate()}")

            elif opcion == 6:
                break

main()
```

```python
import numpy as np


def menu(escoge1):

    print("\n" + "="*50)

    if escoge1 == 1:
        print("        OPERACIONES PARA VECTORES")
        print("="*50)
        print("  1  |  Suma")
        print("  2  |  inverso aditivo")
        print("  3  |  Multiplicacion con un escalar")
        print("  4  |  transpuesta")
        print("  5  |  conjugado")
        print("  6  |  adjunta")
        print("  7  |  producto interno de dos vectores")
        print("  8  |  norma")
        print("  9  |  distancia entre dos vectores")
        print(" 10  |  producto tensor")

    elif escoge1 == 2:
        print("        OPERACIONES PARA MATRICES")
        print("="*50)
        print("  1  |  suma")
        print("  2  |  inversa aditiva")
        print("  3  |  multiplicacion de un escalar")
        print("  4  |  transpuesta")
        print("  5  |  conjugada")
        print("  6  |  Adjunta")
        print("  7  |  producto de dos matrices")
        print("  8  |  revisar si la matriz es unitaria")
        print("  9  |  Revisar si la matriz es hermitiana")
        print(" 10  |  producto tensor")

    elif escoge1 == 3:
        print("   OPERACIONES PARA MATRICES Y VECTORES")
        print("="*50)
        print("  1  |  valores y vectores propios de una matriz")
        print("  2  |  accion de una matriz sobre vector Av")

    print("="*50)


def crear_vector():
    n = int(input("Dimensión del vector (n): "))
    print(f"\nIngrese {n} componentes complejas:\n")
    componentes = []
    for i in range(n):
        print(f"Componente {i+1}")
        re = float(input("  real: "))
        im = float(input("  imag: "))
        componentes.append(complex(re, im))

    orientacion = input("¿Vector columna (c) o fila (f)? [c]: ").lower() or 'c'
    if orientacion == 'f':
        return np.array([componentes])           # shape (1, n)
    else:
        return np.array(componentes)[:, np.newaxis]  # shape (n, 1)


def crear_M(i, j, nombre):

    print("\n" + "="*50)
    print("        CREACION DE MATRIZ")
    print("="*50)

    matriz = []
    for fila in range(i):
        fila_actual = []
        print(f"\nFila {fila+1}")
        print("-"*20)
        for columna in range(j):
            print(f"Elemento [{fila+1},{columna+1}]")
            real = float(input("Parte real:       "))
            imaginario = float(input("Parte imaginaria: "))
            fila_actual.append(complex(real, imaginario))
        matriz.append(fila_actual)

    matriz_np = np.array(matriz)

    print("\nMatriz creada:")
    print("-"*20)
    print(matriz_np)

    return matriz_np


def operacion_matriz(escoge2, m1, m2=None):

    if escoge2 == 1:
        if m2 is not None:
            print("Resultado:")
            print(m1 + m2)
        else:
            print("Se necesitan 2 matrices")

    elif escoge2 == 2:
        print("Resultado:")
        print(-m1)

    elif escoge2 == 3:
        real = float(input("Parte real del escalar: "))
        imag = float(input("Parte imaginaria del escalar: "))
        escalar = complex(real, imag)
        print("Resultado:")
        print(escalar * m1)

    elif escoge2 == 4:
        print("Resultado:")
        print(m1.T)

    elif escoge2 == 5:
        print("Resultado:")
        print(np.conjugate(m1))

    elif escoge2 == 6:
        print("Resultado:")
        print(np.conjugate(m1).T)

    elif escoge2 == 7:
        if m2 is not None:
            if m1.shape[1] == m2.shape[0]:
                print("Resultado:")
                print(np.dot(m1, m2))
            else:
                print("Dimensiones incompatibles para multiplicar")
        else:
            print("Se necesitan 2 matrices")

    elif escoge2 == 8:
        if m1.shape[0] == m1.shape[1]:
            identidad = np.eye(m1.shape[0])
            if np.allclose(np.dot(m1, np.conjugate(m1).T), identidad):
                print("La matriz es unitaria")
            else:
                print("La matriz NO es unitaria")
        else:
            print("Debe ser matriz cuadrada")

    elif escoge2 == 9:
      if m1.shape[0] == m1.shape[1]:
        if np.allclose(m1, np.conjugate(m1).T, atol=1e-10, rtol=1e-8):
            print("La matriz es hermitiana")
        else:
            print("La matriz NO es hermitiana")
      else:
        print("Debe ser matriz cuadrada")

    elif escoge2 == 10:
        if m2 is not None:
            print("Resultado:")
            print(np.kron(m1, m2))
        else:
            print("Se necesitan 2 matrices")

    else:
        print("Invalido")


def operacion_vector(escoge2, v1):

    if escoge2 == 1:
        print("Suma de vectores")
        v2 = crear_vector()
        print("Resultado:")
        print(v1 + v2)

    elif escoge2 == 2:
        print("Inverso aditivo")
        print("Resultado:")
        print(-v1)

    elif escoge2 == 3:
     print("Multiplicacion con un escalar")
     real = float(input("Parte real del escalar:     "))
     imag = float(input("Parte imaginaria del escalar: "))
     escalar = complex(real, imag)
     print("Resultado:")
     print(escalar * v1)

    elif escoge2 == 4:
        print("Transpuesta")
        print("Resultado:")
        print(v1.T)

    elif escoge2 == 5:
        print("Conjugado")
        print("Resultado:")
        print(np.conjugate(v1))

    elif escoge2 == 6:
        print("Adjunta")
        print("Resultado:")
        print(np.conjugate(v1).T)

    elif escoge2 == 7:
        print("Producto interno de dos vectores")
        v2 = crear_vector()
        print("Resultado:")
        print(np.vdot(v1, v2))

    elif escoge2 == 8:
        print("Norma")
        print("Resultado:")
        print(np.linalg.norm(v1))

    elif escoge2 == 9:
        print("Distancia entre dos vectores")
        v2 = crear_vector()
        print("Resultado:")
        print(np.linalg.norm(v1 - v2))

    elif escoge2 == 10:
        print("Producto tensor")
        v2 = crear_vector()
        print("Resultado:")
        print(np.kron(v1, v2))

    else:
        print("invalido")

def Crear_matriz_y_vector():

    print("\nCREAR MATRIZ")
    i = int(input("Numero de filas: "))
    j = int(input("Numero de columnas: "))
    matriz = crear_M(i, j, "MATRIZ")

    print("\nCREAR VECTOR")
    print("El vector debe tener", j, "filas para poder multiplicar A·v")

    vector = []
    for fila in range(j):
        print(f"\nElemento [{fila+1}]")
        real = float(input("Parte real: "))
        imag = float(input("Parte imaginaria: "))
        vector.append([complex(real, imag)])

    vector_np = np.array(vector)

    print("\nVector creado:")
    print(vector_np)

    return matriz, vector_np


def Operacion_matriz_con_vector(opcion, matriz, vector):

    if opcion == 1:
        print("\nValores y vectores propios:")
        if matriz.shape[0] == matriz.shape[1]:
            valores, vectores = np.linalg.eig(matriz)
            print("Valores propios:")
            print(valores)
            print("Vectores propios:")
            print(vectores)
        else:
            print("La matriz debe ser cuadrada.")

    elif opcion == 2:
        print("\nAccion de la matriz sobre el vector (A·v):")
        if matriz.shape[1] == vector.shape[0]:
            resultado = np.dot(matriz, vector)
            print("Resultado:")
            print(resultado)
        else:
            print("Dimensiones incompatibles.")

    else:
        print("Opcion invalida.")

def main():
    while True:
        print("Bienvenido a la lista de operaciones de matriz/vector")
        print("1. si desea vector ")
        print("2. si desea matriz")
        print("3. si desea matriz y el vector")

        escoge1 = int(input("ingrese numero: "))

        if escoge1 == 1:
            vector = crear_vector()
            menu(escoge1)
            escoge2 = int(input("Ingrese opcion: "))
            operacion_vector(escoge2, vector)

        elif escoge1 == 2:
            print("\n" + "="*50)
            print("        GENERADOR DE MATRICES")
            print("="*50)

            cantidad = int(input("Cuantas matrices desea crear? (1 o 2): "))

            if cantidad not in [1, 2]:
                print("Opcion invalida. Se creara 1 matriz por defecto.")
                cantidad = 1

            i = int(input("Ingrese numero de filas: "))
            j = int(input("Ingrese numero de columnas: "))

            m1 = crear_M(i, j, "MATRIZ 1")
            m2 = None

            if cantidad == 2:
                m2 = crear_M(i, j, "MATRIZ 2")

            print("\n" + "="*50)
            print("          MATRICES INGRESADAS")
            print("="*50)
            print("\nMatriz 1:")
            print(m1)

            if m2 is not None:
                print("\nMatriz 2:")
                print(m2)

            menu(escoge1)
            escoge2 = int(input("Ingrese opcion: "))

            if m2 is None:
                operacion_matriz(escoge2, m1)
            else:
                operacion_matriz(escoge2, m1, m2)


        elif escoge1 == 3:

            matriz, vector = Crear_matriz_y_vector()

            menu(escoge1)

            opcion = int(input("Ingrese opcion: "))

            Operacion_matriz_con_vector(opcion, matriz, vector)

        else:
            print("Opcion invalida.")


main()
```
